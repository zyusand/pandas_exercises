### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipo = pd.read_csv(url, sep= "\t")

chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
# chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.strip('$')))
product_more_than_10 = chipo[chipo['item_price'] > 10]['item_name'].unique()
product_more_than_10.shape[0]
```

### Step 5. What is the price of each item? 
###### print a data frame with only three columns item_name choice_description and product_price

```python
item = pd.DataFrame(data=chipo, columns=['item_name', 'choice_description', 'product_price'])
item
```

### Step 6. Sort by the name of the item

```python
item.sort_values('item_name', inplace=True)
item
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
q = item.sort_values('item_price', ascending=False).head(1)
# q['item_name'].values[0]
most_expensive_ordered = item[item['item_name'] == q['item_name'].values[0]]
most_expensive_ordered.groupby('item_name').count()


```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
ordered = chipo[chipo['item_name'] == 'Veggie Salad Bowl'].count()
int(ordered['quantity'])
```

```python
ordered_2 = chipo.groupby('item_name')['item_name'].count()
ordered_2 = ordered_2[ordered_2.index == 'Veggie Salad Bowl']
ordered_2
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
c = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)].groupby('item_name')['item_name'].count()
c
```
