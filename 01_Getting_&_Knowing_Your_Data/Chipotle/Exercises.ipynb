### Step 9. Which was the most-ordered item? 

```python

c = chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False)
c.head(1)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
c = chipo.groupby('item_name').sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

### Step 11. What was the most ordered item in the choice_description column?

```python
# ascending for item by the count of choice_description
groupby_choice_description = chipo.groupby('choice_description').sum()
groupby_choice_description.index[0]
```

### Step 12. How many items were orderd in total?

```python
order_count = chipo.drop_duplicates('item_name')
total_items = len(order_count)
total_items
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo['item_price'].dtype
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.strip("$")))
```

#### Step 13.c. Check the item price type

```python
chipo['item_price'].dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
float(revenue)
```

### Step 15. How many orders were made in the period?

```python
number = chipo.order_id.nunique()
int(number)
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
average_revenue = revenue / number
float(average_revenue)
```

```python
# Solution 2
chipo['revenue_column'] = chipo['quantity'] * chipo['item_price']
average_revenue_1 = chipo.groupby('order_id')['revenue_column'].sum().mean()
average_revenue_1
```
